```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from typing import Literal, Callable
```

```python
data = pd.read_csv("./Clean/DifraccionElectrones_20250906_Voltaje_Diametro_Interno_1_Diametro_Externo_1_Diametro_1_Diametro_Interno_2_Diametro_Externo_2_Diametro_2.dat")
data["error_1"] = data["Diametro_Externo_1"] - data["Diametro_Interno_1"]
data["error_2"] = data["Diametro_Externo_2"] - data["Diametro_Interno_2"]
```

```python
def calcular_error(name_col = "error_1", data : pd.DataFrame = data) -> float:
    return data[name_col].mean()
```

```python
error_1, error_2 = calcular_error(), calcular_error(name_col="error_2")
error_1, error_2
```

```python
inv_sqrt = lambda x: 1/np.sqrt(x)
m1, b1 = np.polyfit(inv_sqrt(data["Voltaje"]), data["Diametro_1"], 1)
m2, b2 = np.polyfit(inv_sqrt(data["Voltaje"]), data["Diametro_2"], 1)
m1, b1, m2, b2
```

```python
def fit_poly(value, line: Literal[1, 2], f: Callable = inv_sqrt ):
    m = m1 if (line == 1) else m2
    b = b1 if (line == 1) else b2
    return m * f(value) + b
```

```python
plt.errorbar(inv_sqrt(data["Voltaje"]), data["Diametro_1"], yerr=error_1,fmt="bo", ecolor="black")
plt.errorbar(inv_sqrt(data["Voltaje"]), data["Diametro_2"], yerr=error_1,fmt="ro", ecolor="black")
plt.plot(inv_sqrt(data["Voltaje"]), fit_poly(data["Voltaje"], line=1))
plt.plot(inv_sqrt(data["Voltaje"]), fit_poly(data["Voltaje"], line=2))
plt.title("PreEliminar: Datos de Diametros respecto a inverso de la raiz del voltaje")
plt.ylabel(r"Diametro ($cm$)")
plt.xlabel(r"Inverso Raiz Voltaje ($V^{-1/2}$)")
```

```python
plt.errorbar(inv_sqrt(data["Voltaje"]), data["Diametro_1"], yerr=data["error_1"],fmt="bo", ecolor="black")
plt.errorbar(inv_sqrt(data["Voltaje"]), data["Diametro_2"], yerr=data["error_2"],fmt="ro", ecolor="black")
plt.plot(inv_sqrt(data["Voltaje"]), fit_poly(data["Voltaje"], line=1))
plt.plot(inv_sqrt(data["Voltaje"]), fit_poly(data["Voltaje"], line=2))
plt.title("PreEliminar: Datos de Diametros respecto a inverso de la raiz del voltaje")
plt.ylabel(r"Diametro ($cm$)")
plt.xlabel(r"Inverso Raiz Voltaje ($V^{-1/2}$)")
plt.savefig("Previo.png")
```
